## Imputation
### Functions
These functions can impute missing categorical and numerical data using various methods such as "mean" and "most_frequent"

```python
# Function to use sklearn to umpute missing values 
def imputeNumericalData(features, df, missing_values, strategy):
    
    # instantiate an imputer object
    imp = Imputer(missing_values=missing_values, strategy=strategy, axis=0)
    
    # iterate through each feature and impute
    for feat in features:
        reshaped = df[feat].reshape(-1, 1)
        df[feat] = imp.fit_transform(reshaped)
        
    return df
    
    
# function to impute categorical using most frequent
def imputeCategorical(features, df):
    # iterate through each feature and impute
    for feat in features:
        df[feat] = df[feat].fillna(df[feat].mode()[0])
        
    return df
```

```python
#impute using most frequent
imputedFeatures_mostFreq = ['facenumber_in_poster', 'aspect_ratio']
df = imputeNumericalData(imputedFeatures_mostFreq, df, np.NaN, 'most_frequent')

# impute categorical using most frequent
imputeCategoricalFeatures_mostFreq = ['language', 'color', 'content_rating']
df = imputeCategorical(imputeCategoricalFeatures_mostFreq, df)

# impute with zero 
df['num_critic_for_reviews'] = df['num_critic_for_reviews'].fillna(0)



df.head()
```

## Prepare Pipe Separated 
Some features in the original dataset had pipe separated values such as "Action|Adventure".<br>
The function below will take these value and put them on their own columns similar to one-hot-encoding.


```python

#creates a new column for item in the original calumn that is seperated by a |
def splitpipe(data, columns):
    
    for column in columns:
        #mySet is a set of every item in the column that is seperated by the pipe
        mySet = set()
        for value in data[column].apply(lambda x: x.split('|')):
            mySet |= set(value)
        for uniqueValue in mySet:
            data[uniqueValue] = data[column].apply(lambda x: 1 if uniqueValue in x else 0)
        data = data.drop(column, 1)
    return data
```

```python
splitlist = ['genres']
df = splitpipe(df, splitlist)
```

## Onehot Encoding Function

```python
def onehotencoding(data, columns):
    for column in columns:
        mySet = set(data[column])
        for uniqueValue in mySet:
            data[uniqueValue] = data[column].apply(lambda x: 1 if x == uniqueValue else 0)
        data = data.drop(column, 1)
    return data
```

```python
ohelist = ['country', 'color', 'aspect_ratio', 'content_rating', 'language']
df = onehotencoding(df, ohelist)
```

## Normalize Function
Normalize to zero mean with unit standard deviation.

```python
def normalize_columns(data, col):
    newdata = pd.DataFrame()
    for c in col:
        newdata[c] =  data[c].apply(lambda x: ((x - np.mean(data[c])) / np.std(data[c])))
        #print(c,np.mean(newdata[c]), np.std(newdata[c]))
    return newdata  
```

## Remove Outliers
This step is necessary as we had a lot movies with really high gross amounts  (Titanic, Avatar) which raised our mean drastically.

```python
def removeOutliers(data, value):
    return data[np.abs(data.gross-data.gross.mean())<=(value*data.gross.std())]

df = removeOutliers(df, 1)
print(df.shape)
```

## Continous to Categorical Function

```python
def cont_to_cat(data, size = 10):
    value_range = data.max()
    n = int(((value_range)/size)+1)
    return data.apply(lambda x: int(x/n))
```
